### Analysis

After cleaning the dataset and from the pie charts and bar charts we observe that the distribution of the mice is relativelty balanced among male and female mice. Also, the number of observed ouse timpoints are quite similar for different drug regiments (slightly higher for Capomulin and Ceftamin). From summary statistic we observe that the tumor volume distribution has smaller variance and standard deviation for Capomulin and Ceftamin. From box plots we can observe that except for Infubinol treatment, the data for other treatments doesn't have outliers. Also, box plots show that the average final tumor volume of the mice under Capomulin and Ceftamin treatments is a lot lower than other treatents. However, from the scatter plot and the regression line calculated for this dataset for weight vs average tumor volume, we observe that the is a strong linear relationshit between the weight of the mice and the size of their tumor for mice in Campolin regimen. If this is also true for other mice under other treatments as well we might conclued that based on this evidence, in general, there is strong positive correlation between mouse weight and size of the tumor.

```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
```

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
complete_df = pd.merge(study_results, mouse_metadata, on='Mouse ID', how='outer')

# Display the data table for preview
complete_df.head()
```

```python
# Checking the number of mice.
len(complete_df['Mouse ID'].unique())
# or
complete_df['Mouse ID'].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = complete_df.loc[complete_df.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
complete_df.loc[complete_df['Mouse ID']=='g989',:]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# cleaned_df = complete_df[complete_df.duplicated(subset=['Mouse ID', 'Timepoint'])==False]
cleaned_df = complete_df[complete_df['Mouse ID'].isin(duplicate_mice)==False]
```

```python
# Checking the number of mice in the clean DataFrame.
cleaned_df['Mouse ID'].nunique()
```

```python
# cleaned_df = complete_df.drop_duplicates(subset=["Mouse ID", "Timepoint"]).nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
means = complete_df['Tumor Volume (mm3)'].groupby(complete_df['Drug Regimen']).mean()
medians = complete_df['Tumor Volume (mm3)'].groupby(complete_df['Drug Regimen']).median()
variances = complete_df['Tumor Volume (mm3)'].groupby(complete_df['Drug Regimen']).var()
stds = complete_df['Tumor Volume (mm3)'].groupby(complete_df['Drug Regimen']).std()
MSEs = complete_df['Tumor Volume (mm3)'].groupby(complete_df['Drug Regimen']).sem()

# Assemble the resulting series into a single summary DataFrame.

summary_stat = pd.DataFrame({"Mean Tumor Volume":means, 
                            "Median Tumor Volume":medians, 
                           "Tumor Volume Variance":variances, 
                           "Tumor Volume Std. Dev.":stds, 
                           "Tumor Volume Std. Err.":MSEs})
summary_stat
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary_aggergation =  cleaned_df.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].agg(['mean', 'median', 'var', 'std', 'sem'])
summary_aggergation
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = cleaned_df.groupby(["Mouse ID"])["Timepoint"].max()
last_timepoint = last_timepoint.reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_data_lasttp = last_timepoint.merge(cleaned_df, on=["Mouse ID", "Timepoint"], how="left")
merged_data_lasttp
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
total_tumor_vol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for trt in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_vol = merged_data_lasttp.loc[merged_data_lasttp["Drug Regimen"] == trt, "Tumor Volume (mm3)"]
    
    # add subset 
    total_tumor_vol.append(tumor_vol)
    
    # Determine outliers using upper and lower bounds
    quartiles = tumor_vol.quantile([.25, .5, .75])
    lowerq = quartiles[.25]
    upperq = quartiles[.75]
    IQR = upperq - lowerq 
    
    lower_bound = lowerq - (1.5*IQR)
    upper_bound = upperq + (1.5*IQR)
    outliers = tumor_vol.loc[(tumor_vol < lower_bound) | (tumor_vol > upper_bound)]
    print(f"{trt}'s potential outlier: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig1, ax1 = plt.subplots(figsize=(6,4))
ax1.set_ylabel('Final Tumor Volume (mm3)',fontsize = 9)
ax1.boxplot(total_tumor_vol, labels=treatments, widths = 0.4, vert=True, flierprops={'marker': 'o', 'markersize': 10, 'markerfacecolor': 'red'})

plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
Capomulin_df = cleaned_df.loc[cleaned_df["Drug Regimen"] == "Capomulin",:]
x = Capomulin_df.loc[Capomulin_df["Mouse ID"] == "l509",:]["Timepoint"]
tumor_size = Capomulin_df.loc[Capomulin_df["Mouse ID"] == "l509",:]["Tumor Volume (mm3)"]
plt.title('Capomulin treatmeant of mouse l509')
plt.plot(x, tumor_size)
plt.xlabel('Timepoint (Days)')
plt.ylabel('Tumor Volume (mm3)')

plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
avg_Capomulin_df = Capomulin_df.groupby(['Mouse ID']).mean()
x = avg_Capomulin_df['Weight (g)']
y = avg_Capomulin_df['Tumor Volume (mm3)']
plt.scatter(x,y)
plt.xlabel("Weight (g)")
plt.ylabel("Averag Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

x = avg_Capomulin_df['Weight (g)']
y = avg_Capomulin_df['Tumor Volume (mm3)']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
reg_val = x * slope + intercept
print(slope)
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x,reg_val,"r-")

plt.scatter(x,y)
plt.xlabel("Weight (g)")
plt.ylabel("Averag Tumor Volume (mm3)")
plt.show()
```
